```python
#strategies
#1.drop
df_cleaned_drop = df.dropna()
print(f"\nDataset shape after dropping missing values: {df_cleaned_drop.shape}")
```

```python
#2.replace
numeric_columns = df.select_dtypes(include=['number'])
df_filled_mean = df.copy()
df_filled_mean[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())
df_filled_mean = df.fillna(df[numeric_columns.columns].mean())
print(df_filled_mean)
```

```python
#importing matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#barchart
plt.figure(figsize=(5,3))
sns.barplot(x='Department', y='Avg Training Score', data=df)
plt.title('Total Avg Training Score per Department')
plt.xlabel('Department')
plt.ylabel('Total Avg Training Score')
plt.xticks(rotation=45)
plt.show()
```

```python
#linechart
plt.figure(figsize=(5,3))
sns.lineplot(x='Department', y='Avg Training Score', data=df)
plt.title('Total Avg Training Score per Department')
plt.xlabel('Department')
plt.ylabel('Total Avg Training Score')
plt.xticks(rotation=45)
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(5,3))
sns.scatterplot(x='No Of Trainings', y='Avg Training Score', data=df)
plt.title('Total Avg Training Score per No Of Trainings')
plt.xlabel('No Of Trainings')
plt.ylabel('Total Avg Training Score')
plt.xticks(rotation=45)
plt.show()
```
