```python
#using strategies to handle dataset
```

```python
#1. using drop strategy
df_cleaned_drop = df.dropna()
print(f"\nDataset shape after dropping missing values: {df_cleaned_drop.shape}")
```

```python
#Check for Missing Values in Dataset after using drop strategy
print(df.isnull().sum())
```

```python
#2.using replace strategy

df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
df['Region'] = df['Region'].fillna(df['Region'].mode()[0])
df['Education'] = df['Education'].fillna(df['Education'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Recruitment Channel'] = df['Recruitment Channel'].fillna(df['Recruitment Channel'].mode()[0])
df['Kpis Met More Than 80'] = df['Kpis Met More Than 80'].fillna(df['Kpis Met More Than 80'].mode()[0])
df['Awards Won'] = df['Awards Won'].fillna(df['Awards Won'].mode()[0])
```

```python
df['No Of Trainings'] = df['No Of Trainings'].fillna(df['No Of Trainings'].mean())
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Previous Year Rating'] = df['Previous Year Rating'].fillna(df['Previous Year Rating'].mean())
df['Length Of Service'] = df['Length Of Service'].fillna(df['Length Of Service'].mean())
df['Avg Training Score'] = df['Avg Training Score'].fillna(df['Avg Training Score'].mean())
```

```python
#Check for Missing Values in Dataset after using replace strategy
print("Missing Values After Filling:")
print(df.isnull().sum())
```

```python
#importing libraries for visualization of dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Bar chart for Total Number of Trainings per Department

plt.figure(figsize=(5, 4))
sns.barplot(x='Department', y='No Of Trainings', data=df, estimator=sum, errorbar=None)
plt.title('Total Number of Trainings per Department')

plt.xlabel('Department')
plt.ylabel('Total Number of Trainings')
plt.xticks(rotation=45)
plt.show()
```

```python
#line chart for Trend of Average Training Score Over Length of Service

avg_training_by_service = df.groupby('Length Of Service')['Avg Training Score'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='Length Of Service', y='Avg Training Score', data=avg_training_by_service, marker='o')

plt.title('Trend of Average Training Score Over Length of Service')
plt.xlabel('Length of Service (Years)')
plt.ylabel('Average Training Score')

plt.tight_layout()
plt.show()
```

```python
# scatterplot for Relationship Between Number of Trainings and Average Training Score

plt.figure(figsize=(10, 6))
sns.scatterplot(x='No Of Trainings', y='Avg Training Score', data=df)

plt.title('Relationship Between Number of Trainings and Average Training Score')
plt.xlabel('Number of Trainings')
plt.ylabel('Average Training Score')

plt.tight_layout()
plt.show()
```

```python
#This scatter plot helps visualize if there is a relationship between the number of trainings and the average training score.
#It shows individual data points where each point represents a record in the dataset, with the x-axis representing the number of trainings and the y-axis representing the training score.
#This visualization helps identify patterns, such as whether additional training leads to higher scores.
```
